#### EDA VARIABLES CATEGÓRICAS

```python
def graficos_eda_categoricos(cat):
    
    #Calculamos el número de filas que necesitamos
    from math import ceil
    filas = ceil(cat.shape[1] / 2)

    #Definimos el gráfico
    f, ax = plt.subplots(nrows = filas, ncols = 2, figsize = (16, filas * 6))

    #Aplanamos para iterar por el gráfico como si fuera de 1 dimensión en lugar de 2
    ax = ax.flat 

    #Creamos el bucle que va añadiendo gráficos
    for cada, variable in enumerate(cat):
        cat[variable].value_counts().plot.barh(ax = ax[cada])
        ax[cada].set_title(variable, fontsize = 12, fontweight = "bold")
        ax[cada].tick_params(labelsize = 12)
```

```python
graficos_eda_categoricos(df.select_dtypes('O'))
```

Conclusiones:

* mayor_edad solo tiene un valor --> eliminarla
* Sobre las imputaciones pendientes de variables categóricas:
  * educacion: imputar por 'Universitaria'

  * implicacion: imputar por 'Alta'

#### EDA VARIABLES NUMÉRICAS

```python
def estadisticos_cont(num):
    #Calculamos describe
    estadisticos = num.describe().T
    #Añadimos la mediana
    estadisticos['median'] = num.median()
    #Reordenamos para que la mediana esté al lado de la media
    estadisticos = estadisticos.iloc[:,[0,1,8,2,3,4,5,6,7]]
    #Lo devolvemos
    return(estadisticos)
```

```python
estadisticos_cont(df.select_dtypes('number'))
```

* Horas quincena solo tiene una valor --> Eliminarla

### ¿Cual es el impacto económico de este problema?

Según el estudio "Cost of Turnover" del Center for American Progress:

* El coste de la fuga de los empleados que ganan menos de 30000 es del 16,1% de su salario

* El coste de la fuga de los empleados que ganan entre 30000-50000 es del 19,7% de su salario

* El coste de la fuga de los empleados que ganan entre 50000-75000 es del 20,4% de su salario

* El coste de la fuga de los empleados que ganan más de 75000 es del 21% de su salario

```python
# Creamos una nueva variable salario_ano del empleado
df['salario_ano'] = df.salario_mes.transform(lambda x: x*12)
df[['salario_mes','salario_ano']]
```

```python
# Calculamos el impacto económico de cada empleado si deja la empresa

#Lista de condiciones
condiciones = [(df['salario_ano'] <= 30000),
               (df['salario_ano'] > 30000) & (df['salario_ano'] <= 50000),
               (df['salario_ano'] > 50000) & (df['salario_ano'] <= 75000),
               (df['salario_ano'] > 75000)]

#Lista de resultados
resultados = [df.salario_ano * 0.161, df.salario_ano * 0.197, df.salario_ano * 0.204, df.salario_ano * 0.21]
                
#Aplicamos select
df['impacto_abandono'] = np.select(condiciones,resultados, default = -999)

df
```

### Y podemos seguir trazando estrategias asociadas a los insights de abandono:

Habíamos visto que los representantes de ventas son el puesto que más se van. ¿Tendría sentido hacer un plan específico para ellos? ¿Cual sería el coste ahorrado si disminuimos la fuga un 30%?

Primero vamos a calcular el % de representantes de ventas que se han ido el año pasado

```python
total_repre_pasado = len(df.loc[df.puesto == 'Sales Representative'])
abandonos_repre_pasado = len(df.loc[(df.puesto == 'Sales Representative') & (df.abandono == 1)])
porc_pasado = abandonos_repre_pasado / total_repre_pasado

porc_pasado
```

Ahora vamos a estimar cuántos se nos irán este año

```python
total_repre_actual = len(df.loc[(df.puesto == 'Sales Representative') & (df.abandono == 0)])
se_iran = int(total_repre_actual * porc_pasado)

se_iran
```

Sobre ellos cuantos podemos retener (hipótesis 30%) y cuanto dinero puede suponer

```python
retenemos = int(se_iran * 0.3)

ahorramos = df.loc[(df.puesto == 'Sales Representative') & (df.abandono == 0),'impacto_abandono'].sum() * porc_pasado * 0.3

print(f'Podemos retener {retenemos} representantes de ventas y ello supondría ahorrar {ahorramos}$.')
```

Este dato también es muy interesante porque nos permite determinar el presupuesto para acciones de retención por departamento o perfil.

Ya que sabemos que podemos gastarnos hasta 37.000$ sólo en acciones específicas para retener a representantes de ventas y se estarían pagando sólas con la pérdida evitada

### Machine Learning

```python
df_ml= df.copy()
```

```python
df_ml.info()
```

```python
# Importar LabelEncoder de sklearn para transformar variables categóricas
from sklearn.preprocessing import LabelEncoder

# Paso 1: Verificar valores nulos antes de la transformación
print("VALORES NULOS ANTES DE LA TRANSFORMACIÓN:")
print(df.isna().sum().sort_values(ascending=False).head())
print(f"\nTotal de valores nulos: {df.isna().sum().sum()}\n")
```

```python
# Paso 3: Identificar todas las columnas categóricas (tipo 'object')
columnas_categoricas = df_ml.select_dtypes(include='object').columns.tolist()
print(f"Variables categóricas encontradas ({len(columnas_categoricas)}): {columnas_categoricas}\n")

# Paso 4: Crear un diccionario para guardar los encoders (útil para futuras transformaciones)
label_encoders = {}

# Paso 5: Aplicar LabelEncoder a cada columna categórica
print("APLICANDO LABEL ENCODER...\n")
for columna in columnas_categoricas:
    # Crear un encoder para esta columna
    le = LabelEncoder()
    
    # Transformar la columna: cada categoría se convierte en un número
    df_ml[columna] = le.fit_transform(df_ml[columna])
    
    # Guardar el encoder en el diccionario
    label_encoders[columna] = le
    
    # Mostrar el mapeo para entender qué número corresponde a cada categoría
    print(f"{columna}: {dict(zip(le.classes_, le.transform(le.classes_)))}")
```

```python
# Paso 6: Verificar que no queden valores nulos después de la transformación
print("\n" + "="*70)
print("VERIFICACIÓN FINAL:")
print("="*70)
print(f"Valores nulos después de la transformación: {df_ml.isna().sum().sum()}")
print(f"Variables categóricas restantes: {len(df_ml.select_dtypes('object').columns)}")
print(f"Variables numéricas totales: {len(df_ml.select_dtypes(['int64', 'float64']).columns)}")
print("\n✓ Todas las variables categóricas han sido transformadas a numéricas")
print("✓ El dataframe 'df_encoded' está listo para usar con sklearn")

# Mostrar el dataframe transformado
df_ml
```

```python
# Importar StandardScaler de sklearn para escalar las variables
from sklearn.preprocessing import StandardScaler

# Paso 1: Verificar el estado actual del dataframe
print("ESTADO ACTUAL DEL DATAFRAME:")
print(f"Forma del dataframe: {df_ml.shape}")
print(f"Tipos de datos:\n{df_ml.dtypes.value_counts()}\n")

# Paso 2: Mostrar estadísticas antes del escalado (para comparar después)
print("ESTADÍSTICAS ANTES DEL ESCALADO:")
print(df_ml.describe().T[['mean', 'std', 'min', 'max']].head(10))
print()

# Paso 3: Crear una copia para el escalado
df_scaled = df_ml.copy()

# Paso 4: Separar la variable objetivo (abandono) del resto
# Es importante NO escalar la variable objetivo si es binaria (0/1)
target = 'abandono'
X_columns = [col for col in df_scaled.columns if col != target]

print(f"Variables a escalar: {len(X_columns)}")
print(f"Variable objetivo (no se escalará): {target}\n")
```

```python
# Paso 5: Crear el escalador StandardScaler
# StandardScaler transforma cada variable a media=0 y desviación estándar=1
scaler = StandardScaler()

# Paso 6: Aplicar el escalado a todas las variables excepto la variable objetivo
df_scaled[X_columns] = scaler.fit_transform(df_scaled[X_columns])

# Paso 7: Verificar las estadísticas después del escalado
print("="*70)
print("ESTADÍSTICAS DESPUÉS DEL ESCALADO:")
print("="*70)
print(df_scaled[X_columns].describe().T[['mean', 'std', 'min', 'max']].head(10))
print()

# Paso 8: Verificación final
print("="*70)
print("VERIFICACIÓN FINAL:")
print("="*70)
print(f"✓ Todas las variables tienen media cercana a 0")
print(f"✓ Todas las variables tienen desviación estándar = 1")
print(f"✓ Las variables ahora tienen la misma escala e importancia")
print(f"✓ La variable objetivo '{target}' se mantiene sin cambios")
print(f"\nValores nulos en df_scaled: {df_scaled.isna().sum().sum()}")

# Mostrar el dataframe escalado
df_scaled
```
