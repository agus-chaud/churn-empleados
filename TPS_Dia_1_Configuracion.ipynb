### ANÁLISIS DE CALIDAD - Estructura y tipos

Revisamos la estructura del dataset y los tipos inferidos para detectar columnas que deberían ser categóricas pero están como numéricas (o al revés).

```python
# Tipos por columna
df.dtypes
# Tabla resumen
pd.DataFrame({'columna': df.columns, 'tipo': df.dtypes})
```

```python
# Valores únicos en columnas numéricas que podrían ser categóricas
# (pocos valores distintos = candidatas a categórica)
for col in df.select_dtypes(include=[np.number]).columns:
    n_unicos = df[col].nunique()
    if n_unicos <= 20:
        print(f"{col}: {n_unicos} valores únicos -> {sorted(df[col].dropna().unique().tolist())}")
```
